```python
import pandas as pd
import psycopg2
```

```python
big_cheddar_csv = "Resources/big_cheddar.csv"
big_cheddar_df = pd.read_csv(big_cheddar_csv)
big_cheddar_df.head(15)
```

```python
clean_cheese_df = big_cheddar_df
```

```python
bigger_cheddar_df = pd.read_csv('Resources/bigger_cheddar.csv')
bigger_cheddar_df.head()
```

```python
bigger_cheddar_df = bigger_cheddar_df[['Date','Weighted Price', 'Weighted Price adjusted to 38% moisture', 'Sales', 'Moisture Content']].copy()
bigger_cheddar_df.head()
```

```python
big_c = bigger_cheddar_df
```

```python
clean_cheese_df = pd.DataFrame({"40 LB CHEESE":big_c.groupby(['Date', 'Sales'])["Weighted Price"].mean()})
```

```python
clean_cheese_df.head()
```

```python
file_df = pd.read_csv('Resources/file.csv')
file_df.head()
```

```python
file_df = file_df[["","",""]]
file_df.head()
```

```python
file_df = file_df.rename(columns{"","",""})
```

```python
file_df_new = pd.dataFrame({"":file_df.groupby(["",""])[""].mean(),
      "40 LB CHEESE":file_df.groupby(["Date","Sales"]["Weighted Price"].sum()})                     })
```

```python
file_df_new.head()
```

```python
#MERGING DATA
merged = pd.merge(clean_cheese_df,file_df_new, on=["",""],how="inner")
```

```python
merged.head()
```

```python
merged2 = pd.merge(merged,file_df_new, on=["",''],how="inner")
```

```python
merged2.head()
```

```python
merged_final = merged2.reset_index()
```

```python
merged_final.head()
```

```python
merged_final = merged_final[["","","","","","","",""]]
```

```python
from sqlalchemy import create_engine
import pymysql
pymysql.install_as_MySQLdb()
```
